#### Train new word2vec

```python
tweet_tokens = []

with open('/data/rumor_detection/rumor_detection/data/interim/source_tweets.txt', 'r') as file:
    for line in file:
        tweet_tokens.append(line.split(' '))
```

```python
model_1 = gensim.models.Word2Vec(tweet_tokens, size=300, window=10, min_count=5, iter=10, workers=1)
```

```python
model_2 = gensim.models.Word2Vec(tweet_tokens, size=300, window=10, min_count=5, iter=10, workers=1)
```

```python
model_1['play'] - model_2['play']
```

```python
embed_model_file = '/data/rumor_detection/rumor_detection/pretrain_models/cv/sydneysiege-all-rnr-threads.txt'
embed_model_file_update = '/data/rumor_detection/rumor_detection/pretrain_models/cv/update_sydneysiege-all-rnr-threads.txt'
```

```python
embed_dict = {}
with open(embed_model_file, 'r') as file:
    for line in file:
        line_arr = line.split(' ', 1)
        embed_dict[line_arr[0]] = line
```

```python
with open(embed_model_file_update + ".fix", 'w') as writer:
    with open(embed_model_file_update, 'r') as file:
        for line in file:
            line_arr = line.split(' ', 1)
            if line_arr[0] in embed_dict:
                writer.write(embed_dict[line_arr[0]])
            else:
                writer.write(line)
```
